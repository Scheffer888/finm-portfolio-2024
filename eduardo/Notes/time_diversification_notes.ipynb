### Log Returns

| notation        | description                                 |
|-----------------|---------------------------------------------|
| $R_t$           |                          return rate, $(t - 1 \text{ to } t)$       |
| $r_{t, t+h}$    | $\left( \prod_{\tau=1}^{h} 1+ R_{t+\tau} \right) - 1$ cumulative return rate, from $t$ to $t + h$ |
| $r_t$           | $\log(1 + R_t)$         log return from $t - 1$ to $t$             |
| $r_{t, t+h}$    | $\log(1 + R_{t, t+h})$  log cumulative return from $t$ to $t + h$ |

### Log returns

Let $r$ denote the log-return: $$r_t \equiv \ln(1 + R_t)$$

If $X \sim N(\mu, \sigma)$, then: $$Y = e^X \Rightarrow \ln Y = X \sim N(\mu, \sigma)$$

This means that $Y$ is lognormally distributed.

For returns, we have: $$R_t = e^r_t - 1$$

Using the stock price relation: $$S_t (R_t + 1) = S_t e^r_t$$

This implies that: $$\ln(1 + R_t) = \ln e^r_t$$

And therefore: $$\ln \left( \frac{S_{t+1}}{S_t} \right) = r_t$$


Log returns are particularly useful when dealing with compounding returns across different time horizons.

The cumulative log return over a horizon $t$ to $t + h$ can be expressed as: $$r_{t, t+h} \equiv \log(1 + R_{t, t+h}) = \sum_{i=1}^{h} r_{t+i}$$

Thus, the **cumulative return from $t$ to $t + h$ is just the sum of one-period returns.**


### Cumulative Return Risk and Autocorrelation

Log cumulative returns are simply a portfolio of single-period returns.

- Define $ \mu $ and $ \sigma^2 $ as
  $$\mathbb{E} [ r_{t, t+1} ] = \mu,$$
  $$\text{var} [ r_{t, t+1} ] = \sigma^2$$

- For the $ h $-period log return,
  $$\mathbb{E} [ r_{t, t+h} ] = h \mu $$
  $$\text{var} [ r_{t, t+h} ] = \sum_{j=1}^{h} \sum_{i=1}^{h} \text{cov} [ r_{t+i}, r_{t+j} ]$$
- If covariances are zero, then (log returns + i.i.d.): 
  $$\text{var} [ r_{t, t+h} ] = h \sigma^2$$

--- 
### Autocorrelation Models

As seen in the previous slide,

- The variance of cumulative returns, $ r_{t, t+h} $, depends critically on the auto-covariance of the return series, denoted as
  $$ \text{cov} [ r_t, r_{t+i} ], \text{ or } \sigma_{t, t+i}$$

- Specifying a form for the autocorrelations, $ \text{corr} [ r_t, r_{t+i} ] $, is equivalent.

- A model of autocorrelations uniquely specifies a linear time series model.

#### AR(1) Models

Autoregressive (AR) models are among the most popular in time-series statistics.

Consider the AR(1) model,

$$ \text{cov} [ r_t, r_{t+i} ] = \rho^i \sigma^2 $$
$$ \text{corr} [ r_t, r_{t+i} ] = \rho^i $$

With AR models, covariances are easy to scale over time.

--- 

### Diversification and Cumulative Returns

Mean returns scale linearly in horizon $ h $,
$$\mathbb{E} [ r_{t, t+h} ] = h \mu$$

But scaling of variance depends on correlation,

- $ \rho = 1 $: Std. Dev. is linear in cumulation: $ \text{std} [ r_{t, t+h} ] = h \sigma $
- $ \rho = 0 $: Variance is linear in cumulation: $ \text{var} [ r_{t, t+h} ] = h \sigma^2 $
- $ \rho = -1 $: The return is riskless: $ \text{var} [ r_{t, t+h} ] = 0 $

*Time diversification with less than 1 correlation: by holding on to an asset through many periods.*
- Related to the diversification of a portfolio across assets.

#### Example: Riskless Bond

At time $ t = 0 $, consider a bond with riskless payout at $ t = 10 $. The yield of the bond at $ t = 0 $ is 5%.

- The 10-year cumulative return, $ r_{0,10} $, is riskless and equals $ 5\% \times 10 $.
- At any intermediate time, $ t $, such that $ 0 < t < 10 $, the bond price is uncertain.
- Thus, the intermediate returns, $ r_{0,t} $ and $ r_{t,10} $, are uncertain.
- However, $ r_{0,10} = r_{0,t} + r_{t,10} $.
- So if $ r_{0,t} $ is unexpectedly high, then $ r_{t,10} $ must be relatively low, such that the riskless return $ r_{0,10} $ ends up at $ 5\% \times 10 $.

### Negative Serial Correlation in Bonds

Thus, riskless bonds should have negatively autocorrelated returns:
$$\text{corr} (r_t, r_{t+1}) < 0$$

And if the bond matures at $ T $, then for any $ 0 < h < T - t $,
$$\text{corr} (r_{t, t+h}, r_{t+h, T}) = -1$$

Default-free bonds are safer in the long-run, with $\text{var} [r_{t, T}] = 0$.

### Cumulative Sharpe Ratios in AR(1) Model

Consider again the cumulative return, $r_{t, t+h}$.

- For $ \rho = 1 $,
  $$\text{SR} (r_{t, t+h}) = \text{SR} (r_t)$$

- For $ |\rho| < 1 $,
  $$\text{SR} (r_{t, t+h}) > \text{SR} (r_t)$$

- For $ \rho = 0 $,
  $$\text{SR} (r_{t, t+h}) = \sqrt{h} \, \text{SR} (r_t)$$

*Note:* Time diversification.

### Mean Annualized Returns

The annualized mean (log) return on the $h$-period investment, $r_{t, t+h}$, is
$$\frac{r_{t, t+h}}{h} = \frac{\sum_{i=1}^{h} r_{t+i}}{h}$$
$\rightarrow$ *uncertainty in cumulative returns remains high.*

This is just the usual sample estimate of the mean of one-period returns, $\bar{r}$, based on a sample-size of $h$!

- For any $\rho$, $$\mathbb{E} [\bar{r}] = \mu$$ --> *The biggest uncertainty that casts doubt on the result.*

- For $\rho = 0$, $$\text{var} [\bar{r}] = \frac{\sigma^2}{h}$$

- So as the investment horizon gets large, the mean annualized return, $\bar{r}$, converges to the true annual mean return, $\mu$.
- Even if $\rho \neq 0$, we can still conclude that $\bar{r} \rightarrow \mu$ (Law of large numbers still holds).

However: this might not happen for some reasons:
1. Actual stock returns are not coming from some stochastic generating process based around a static mean-based distribution of $\mu$.
  - Over 100 years, companies like GE and even indices like the S&P 500 have changed drastically, reflecting shifts in technology, investor demographics, politics, and tax environments.
  - Returns aren’t generated from a simple IID process, and companies evolve, introducing risks like bankruptcy. So, the static model assumption doesn’t hold.
2. It is not clear what $\mu$ is.
  - So, even if it were static, it would take me years and years and years to have a good estimate of it.
3. Do investors really care in the long run about the average annual return? --> No, they care about the cumulative return.
  - Cumulative return is what matters, and it is not riskless.
  - When we calculate the cumulative return, we see that the variance is still growing with the horizon (not going to zero as the horizon increases).

## Time Diversification

**Time-diversification** refers to the idea that mean annualized return becomes riskless for large investment horizons.

- True, as horizon increases, the variance of the annualized return goes to zero.
- However, the variance of the cumulative return, $r_{t, t+h}$, is still growing.

So-called time diversification depends on the risk one is measuring.


Still, **Time Diversification has benefit**:
- Even though the mean can change over time and returns are not perfectly i.i.d, if the mean is positive, your cumulative return has a growing expected value as the horizon increases.
- Also, as long as the relationship between $r_{t-1}$ and $r_{t}$ is not 1 (which holds in reality), the standard deviation of the cumulative return scales sublinearly.
- Thus, the first moment of the series (returns) scales linearly (even if the mean is changing) and the standard deviation scales sublinearly. As a consequence, your Sharpe is an increasing function with respect to the horizon.
- An increase in the horizon still leads to time diversification as long as the autocorrelation of returns is not 1 (which holds true).

*Example:* a strategy expected to have 0.3 Sharpe for the next year will have a Sharpe of 1.5 for the next 25 years (considering the basic assumptions hold). Consider the plot showing the cumulative return over the next year (blue) and the next 25 years (red):
- The expected cumulative return is bigger in the 25-year curve. On the other hand, the volatility for the cumulative return of the 25 is also bigger.
- Nonetheless, because the volatility increases sublinearly, the CDF of the red curve is to the "right" of the CDF of the blue curve. Also, because the volatility increases sublinearly the Sharpe of the red cumulative return is bigger.

**When does the "buy and hold" strategy actually make sense?**
In the lens of a quant, an investor that does "buy and hold", if the investor has pertained of a good rationale, is holding the asset because they keep thinking that the expected return is still attractive considering their portfolio regardless of movements in the market.
Nonetheless, if those investors are not willing to adjust positions when new information is available, they lose great opportunities to adjust for risk. This means that, after movements in the market, your expected return for many of the available assets should change as well as their risk metrics: making your positions 100% static is ignoring the new information.
Why can it still work? While those investors do not do a good job adjusting risk (if they only "hold" regardless of movements), some of them have above-average capacity to define expected returns of assets.
