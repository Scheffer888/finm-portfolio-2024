### 3.1. Lagged Regression

The lagged regression is given by: (1)
$$r^{SPY_t} = \alpha^{SPY,X} + (\beta^{SPY,X})' X_{t-1} + \epsilon^{SPY,X_t}$$  

- This can be seen as a linear approximation of a conditional expectation. A way to model how X should predict SPY.
- Epsilon may have a lot of variance relative to X (R square is very small). It still doesn't mean that X is not a predictor, it can just mean that it has a weak effect (doesn't explain much of the y variance).
- For publicly traded investments, an R^2 of 10% is already very high. Probably will get something closer to 1%

#### (a) Estimate and report R², α, and β for:
- X as a single regressor: the dividend-price ratio
- X as a single regressor: the earnings-price ratio
- X as three regressors: the dividend-price ratio, the earnings-price ratio, and the 10-year yield  

Although we are estimating t+1, since these signals have high serial correlation, they might work better at longer horizons.


```python
lag = 1

# lag the independent variable
signals_lagged = signals.shift(lag).dropna()
# align the data frames to the same subset of dates
signals_lagged, spy = signals_lagged.align(excess_returns[['SPY']], join='inner',axis=0)

model_map = {'DP': ['SPX DVD YLD'],
             'P/E': ['SPX P/E'],
             'DP + P/E + 10Y': ['SPX DVD YLD', 'SPX P/E', 'TNote 10YR']}

regression_df = pd.DataFrame({})

for model_name in model_map.keys():
    X = signals_lagged[model_map[model_name]]
    regression = pm.calc_regression(Y=spy,
                                    X=X,
                                    annual_factor=12,
                                    intercept=True,
                                    keep_indexes=['Annualized Alpha', 'Beta (SPX DVD YLD)', 'Beta (SPX P/E)', 'Beta (TNote 10YR)', 'R-Squared']
                                    )
    regression = regression.rename(columns={'SPY': f'SPY ({model_name})'})
    regression_df = pd.concat([regression_df, regression], axis=1)

print(f'Lagged regressors ({lag}-period):')
display(regression_df.fillna(0))
```

#### (b) For each of the three regressions, utilize the resulting forecast in a trading strategy.

You should now have the trading strategy returns, $r^x$, for each of the forecasts. For each strategy, estimate:

# Section 4: Out-of-Sample Forecasting

Reconsider the problem above, of estimating (1) for $x$. The reported $R^2$ was the in-sample $R^2$—it examined how well the forecasts fit in the sample from which the parameters were estimated.

**In particular, focus on the case of using both dividend-price and earnings-price as signals.**

Let’s consider the out-of-sample $R^2$. To do so, we need to do the following:

- Start at $t = 60$.

- Estimate the lagged regression only using data through time $t$. (1)
$$r^{SPY_t} = \alpha^{SPY,X} + (\beta^{SPY,X})' X_{t-1} + \epsilon^{SPY,X_t}$$  

- Use the estimated parameters of (1), along with $x_{t+1}$ to calculate the out-of-sample forecast for the following period, $t + 1$:
  $$
  \hat{r}_{t+1}^{SPY} = \hat{\alpha}_t^{SPY,x} + (\beta^{SPY,x})' x_t
  $$

- Calculate the $t + 1$ forecast error,
  $$
  e_{t+1}^{\text{forecast}} = r_{t+1}^{SPY} - \hat{r}_{t+1}^{SPY}
  $$

- Move to $t = 61$, and loop through the rest of the sample.

You now have the time-series of out-of-sample prediction errors, $e^x$.

Calculate the time-series of out-of-sample prediction errors $e^{\text{null}}$, which are based on the null forecast:
  $$
  \bar{r}_{t+1}^{SPY} = \frac{1}{t} \sum_{i=1}^t r_i^{SPY}
  $$
  $$
  e_{t+1}^{\text{null}} = r_{t+1}^{SPY} - \bar{r}_{t+1}^{SPY}
  $$

#### 4.1 Report the out-of-sample $R^2$**:
   $$
   R^2_{OOS} \equiv 1 - \frac{\sum_{i=61}^T (e_i^{\text{forecast}})^2}{\sum_{i=61}^T (e_i^{\text{null}})^2}
   $$

   Note that unlike an in-sample $R^2$, the out-of-sample $R^2$ can be anywhere between $(-\infty, 1]$.

- Did this forecasting strategy produce a positive OOS $R^2$?

```python
# Use rolling window to calculate the out-of-sample performance

regressors = signals[['SPX DVD YLD', 'SPX P/E']]
oos_ep_dp_regres_stats = pm.calc_replication_oos_perf(y=excess_returns['SPY'],
                                         X=regressors,
                                         annual_factor=12,
                                         intercept=True,
                                         rolling=False,
                                         window_size=60,
                                         lag_periods=1,
                                         return_model_param=True
                                         )
oos_ep_dp_regres_stats
```
